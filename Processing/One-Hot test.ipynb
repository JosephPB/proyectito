```python
import numpy as np
import os
import sklearn.preprocessing
```

```python
def OneHot(a):
    height,width = a.shape
    a = a.flatten()
    a = a.astype(int)
    label_binarizer = sklearn.preprocessing.LabelBinarizer()
    label_binarizer.fit(range(max(a)+1))
    b = label_binarizer.transform(a)
    b = b.reshape((height,width,a.max()+1))
    return b
```

```python
a = np.array([[1,2,3,4],[3,4,1,2]])
OneHot(a)
```

```python
dir = '/media/sf_TestImageDataBase/training/labels/'
```

```python
for file in os.listdir(dir):
    if file.endswith(".npy"):
        directory = os.path.join(dir, file)
        inpt = np.load(directory)
        mat = OneHot(inpt)
        out_name = directory[:-4]    
        np.save('{}'.format(out_name + 'onehot'),mat)
```
