```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
import cv2
import rawpy
import imageio
```

```python
tiff_file = rasterio.open("Wrist_nc.tif")
tiff_image = tiff_file.read(1)
jpeg_image = cv2.imread("Wrist_nc.jpg")
label_image = cv2.imread("Wrist_label.jpg")
print(tiff_image)
```

```python
plt.imshow(tiff_image, cmap = 'gray')
plt.show()
```

```python
plt.imshow(jpeg_image)
```

```python
plt.imshow(label_image)
```

```python
print(jpeg_image.shape)
print(tiff_image.shape)
print(label_image.shape)
```

```python
#resize
jpeg_resized = cv2.resize(jpeg_image, (200,200))
tiff_resized = cv2.resize(tiff_image, (200,200))
label_resized = cv2.resize(label_image, (200,200))
```

```python
print("jpeg")
plt.imshow(jpeg_resized, cmap = 'gray')
plt.show()
print("tiff")
plt.imshow(tiff_resized, cmap = 'gray')
plt.show()

print("jpeg")

plt.imshow(jpeg_resized, cmap = 'gray')
plt.imshow(label_resized, alpha=0.2)
plt.show()
print("tiff")
plt.imshow(tiff_resized, cmap = 'gray')
plt.imshow(label_resized, alpha=0.2)

plt.show()
```

```python
#check upscaling
label_upscaled = cv2.resize(label_resized, (tiff_image.shape[0], tiff_image.shape[1]))
plt.imshow(tiff_image, cmap = 'gray')
plt.imshow(label_upscaled, alpha = 0.2)
plt.show()
```
